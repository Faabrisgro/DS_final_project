## 1. Variables del dataset
### Datos de los clientes del banco:
1.  Edad: Entero.
2.  Trabajo: Tipo de trabajo (categrico).
3.  Estado civil: Casado, divorciado, soltero. (divorciado aplica tambin para viudo/a). 
4.  Nivel de educacin: Categrico (categrico con diferentes niveles de educacin).
5.  Default: Tiene credito en default? (categrico "Yes" o "No").
6.  Hipoteca: Tiene hipoteca? (categorico "No","Yes",'desconocido').
7.  Prstamos personal: Tiene prstamos personal? (categorico "No","Yes","Desconocido").
### Relacionado con el ltimo conctacto con la campaa activa:
8.  Contacto: Tipo de contacto (categorico 'cellular' o 'telephone').
9.  Mes: ltimo mes en el que se contact al cliente (categrico, enero, feb, mar, etc.) 
10.  Dia: ltimo da en el que se contact al cliente (categrico, lunes, martes, mircoles, jueves, viernes) 
11.  Duracin: Duracin del ltimo contacto en segundos (entero). No se tendr en cuenta para el desarrollo del modelo predicitivo. 
### Otros atributos o variables:
12.  Campaa: Nmero de contactos realizados durante la campaa para cada cliente (entero e incluye el timo contacto).
13.  Dia_p: Nmero de das que pasaron despus de que el cliente fue contactado entre campaa y campaa (Entero. 999 significa que el cliente no fue previamente contactado).
14.  Previo: Nmero de contactos realizados antes de esta campaa para cada cliente (Entero)
15.  Previo_result: Resultado de la campaa de marketing previa (categorico: 'fracaso','no existente','exito')
### Variables sociales y de contexto econmico:
16.  emp.var.rate: employment variation rate - Ratio de variacin de empleamiento, es un indicador cuartil.
17.  cons.price.idx: consumer price index - Precio del consumidor, es un indicador mensual.
18.  cons.conf.idx: consumer confidence index - Confianza del consumidor, es un indicador mensual.
19.  euribor3m: euribor 3 month rate - Euribor, ratio de 3 meses, indicador diario.
20.  nr.employed: number of employees - Nmero de empleados, indicador cuartil. 

### Variable a predecir:
21.  y - El cliente ha realizado un plazo fijo? Binario: S o No. 

## 2. Abstracto 

Los datos estn relacionados con campaas de marketing directo (llamadas telefnicas) de una institucin bancaria portuguesa. El objetivo es predecir si el cliente realizar un plazo fijo en el futuro. 

Esto se debe a que el banco quiere maximizar el retorno de inversin que realiza en sus campaas de marketing directo. Si logramos identificar y predecir aquellos consumidores con mayor capacidad de realizar un plazo fijo. Le podramos ahorrar mucho tiempo y dinero al departamento de marketing directo y telemarketing. Ya que sus llamadas estaran mejor segmentadas y tendran mayor probabilidad de ser exitosas gracias al anlisis de datos y el modelo que se necesita implementar.

Me motiva mucho el anlisis de este dataset ya que me permite manejar variables que entiendo por haber estudiado marketing y por otro lado, este dataset podra permitirme la implementacin de dos tipos de modelos, supervisado si lo que queremos es lograr predecir la posibilidad de que un consumidor haga un plazo fijo, el actual objetivo. Y una posible alternativa, la identificacin y agrupamiento de sectores de consumidores con mayor y menos probabilidad de realizar un plazo fijo.

1. Las personas que no tienen hipotecas ni prstamos personales tienen mayor posibilidad de realizar un plazo fijo.

3. Las personas con mayor nivel de educacin tienen ms probabilidad de realizar un plazo fijo.

### Comencemos a probar o eliminar nuestras hiptesis

```python
df.columns
```

```python
df = df.rename(columns={"y": "p_fijo"}) #cambio de nombre para en el EDA identificar mejor s hizo o no un depsito de plazo fijo.
df
```

```python
grouped = df.groupby(['loan','p_fijo', 'balance']).sum()
grouped
```

```python
sns.stripplot(data=grouped, x='loan', y='balance', hue = 'p_fijo').set(title='Distribucin del balance de clientes con prstamos y sin prstamos')
```

> Para nuestra 1 Hiptesis nos interesan los puntos en color amarillo. Es decir, personas que s hicieron un plazo fijo

```python
rel= sns.catplot(data=grouped, x='loan', y='balance', col= 'p_fijo')
rel.fig.suptitle('Balance de clientes')
```

> Ac podemos ver mejor las personas que s hicieron un plazo fijo. Se distribuy en dos, los que s y los que no tienen prstamos personales

En conclusin, es verdad nuestra hiptesis de que los clientes que han realizado un plazo fijo, en su gran mayora lo hicieron sin crditos o prstamos personales. Sin embargo, llama la atencin el gran nmero de clientes que sin un prstamo no han hecho un depsito. Eso puede deberse a otras variables.

> Ahora veamos nuestra hipstesis sobre las edades 

```python
edades = df.groupby(['age']).value_counts()
edades = pd.DataFrame(edades)
```

```python
edades = edades.sort_index(ascending = True)
```

```python
edades.reset_index(inplace=True)
```

```python
edades= edades.sort_values(by='age')
```

```python
plt.figure(figsize = (12,4))
plt.subplot(1,3,1)
sns.barplot(x= edades['age'][0:3050], y= edades['balance'][0:3050], hue= edades['p_fijo'][0:3050])

plt.subplot(1,3,2)
sns.barplot(x= edades['age'][3050:26008], y= edades['balance'][3050:26008], hue= edades['p_fijo'][3050:26008])

plt.subplot(1,3,3)
sns.barplot(x= edades['age'][26009:30000], y= edades['balance'][26009:30000], hue= edades['p_fijo'][26009:30000])
```

```python
plt.figure(figsize = (12,4))
plt.subplot(2,3,1)
sns.barplot(x= edades['age'][30000:40000], y= edades['balance'][30000:40000], hue= edades['p_fijo'][30000:40000])

plt.subplot(2,3,2)
sns.barplot(x= edades['age'][40000:44000], y= edades['balance'][40000:44000], hue= edades['p_fijo'][40000:44000])

plt.subplot(2,3,3)
sns.barplot(x= edades['age'][44000:44500], y= edades['balance'][44000:44500], hue= edades['p_fijo'][44000:44500])

plt.subplot(2,3,4)
sns.barplot(x= edades['age'][44500:44600], y= edades['balance'][44000:44600], hue= edades['p_fijo'][44000:44600])

plt.subplot(2,3,5)
sns.barplot(x= edades['age'][44600:44800], y= edades['balance'][44600:44800], hue= edades['p_fijo'][44600:44800])
```
